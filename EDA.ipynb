```python
df['Population'].mean() # ÌèâÍ∑† = 6162876.3

import scipy.stats

scipy.stats.trim_mean(df['Population'], 0.1) # = Ï†àÏÇ¨ÌèâÍ∑† 4783697.125
df['Population'].median() # Ï§ëÍ∞ÑÍ∞í = 4436369.5
```
